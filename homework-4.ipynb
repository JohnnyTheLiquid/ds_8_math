Придумайте функции со следующими свойствами (в a) b) c) область определения X любая, какая вам удобна, главное - функция с нужными свойствами):

```python
import numpy as np
import matplotlib.pyplot as plt
from math import cos, sin, pi, log
```

a) Область определения от минус бесконечности до плюс бесконечности
$$y = sin(x)$$

```python
def func1(x):
    y = sin(x)
    return y

x1 = np.linspace(-4*pi, 4*pi, 10000)
y1 = [func1(x1[i]) for i in range(len(x1))]

plt.plot(x1, y1)
plt.title('График функции')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

b) Кубическая парабола имеет 1 локальный минимум и 1 локальный максимум, соответственно, еще точку перегиба:
$$y = x^{3} + 5x$$

```python
def func2(x):
    y = pow(x,3) + 3 * pow(x,2)
    return y

x1 = np.linspace(-5, 3, 100)
y1 = [func2(x1[i]) for i in range(len(x1))]

plt.plot(x1, y1)
plt.title('График функции')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

c) Парабола четвертой степени вида 
$$y = x^{4} - 2x^{2} + 10$$ имеет 3 точки экстремума

```python
import numpy as np
import matplotlib.pyplot as plt

def func3(x):
    y = pow(x,4) - 2 * pow(x,2) + 10
    return y


x1 = np.linspace(-2, 2, 50)
y1 = [func3(x1[i]) for i in range(len(x1))]

plt.plot(x1, y1)
plt.title('График функции')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

### Задание 3:

Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$


```python
from scipy.misc import derivative

def func(x):
    y = pow(x,5) + 4 * sin (2 * x) + cos (3 * x + 3)
    return y

y3x = derivative(func, 1.0, dx=1e-6, n=3, order=5)
y4x = derivative(func, 1.0, dx=1e-6, n=4, order=5)
print(y3x)
print(y4x)
```
