Найти множество параметров (strategy=..., maxiter=..., ..., updating=...), которые требуют минимального количества итераций.

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e
```

```python
bounds = [(-10, 10), (-10, 10)]
result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
def evol_strat (func, var):
    bounds = [(-10, 10), (-10, 10)]
    result = differential_evolution(ackley, bounds, strategy=var, seed=42)
    return result.nit
```

```python
var = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp', 'best2exp', 'rand2exp', 'randtobest1bin'
,'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']

Y=[]

for i in var:
    Y.append(evol_strat(ackley, i))
```

```python
plt.plot(var, Y)
plt.title('Dependance of iterations on "strategy"')
plt.xticks(rotation=55)
plt.xlabel('Strategy')
plt.show()
```

```python
def evol_maxiter (func, var):
    bounds = [(-10, 10), (-10, 10)]
    result = differential_evolution(ackley, bounds, maxiter=var, seed=42)
    return result.nit
```

```python
var = np.linspace(0,100,11)

Y=[]

for i in var:
    Y.append(evol_maxiter(ackley, int(i)))
```

```python
plt.plot(var, Y)
plt.title('Dependance of iterations on "maxiter"')
plt.xlabel('maxiter')
plt.show()
```
